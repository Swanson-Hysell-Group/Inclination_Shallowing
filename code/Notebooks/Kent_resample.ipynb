# Notebook for demonstrating the new `pmag.kentdev` and `ipmag.kentrot` function

```python
import numpy as np

import pmagpy.pmag as pmag
import pmagpy.ipmag as ipmag
from pmagpy import pmagplotlib
import random
```

## demo `pmag.kentdev'`

```python
decs, incs = pmag.kentdev(1000, 500, 5000)
```

```python
m = ipmag.make_orthographic_map(0,90)

ipmag.plot_vgp(m,decs, incs, color='gray', edge='none', alpha=0.2)
```

## demo `ipmag.kentrot`

```python
Cutface_kent_stats = {'dec': 184.28941174976043,
                     'inc': 28.054950595968013,
                     'n': 500000,
                     'Zdec': 297.811041170858,
                     'Zinc': 36.82790528859862,
                     'Edec': 67.34980125363006,
                     'Einc': 40.36777571686788,
                     'Zeta': 6.594599514092355,
                     'Eta': 1.8303189378105758,
                     'R1': 0.9988149139276513,
                     'R2': 0.002026226898911236}

Jacobsville_kent_stats = {'dec': 186.32613004793535,
                     'inc': -14.199196016692765,
                     'n': 100000,
                     'Zdec': 109.25383075442295,
                     'Zinc': 41.48280413757106,
                     'Edec': 261.6376932010964,
                     'Einc': 45.06093458505846,
                     'Zeta': 2.972678691710591,
                     'Eta': 3.308247732230898,
                     'R1': 0.998,
                     'R2': 0.001}
```

```python
CF_decs, CF_incs = ipmag.kentrot(Cutface_kent_stats,2000, di_block=0)
m = ipmag.make_orthographic_map(200,20)

ipmag.plot_vgp(m, CF_decs, CF_incs, color='gray', edge='none', alpha=0.2)
```

```python
J_decs, J_incs = ipmag.kentrot(Jacobsville_kent_stats,2000, di_block=0)
m = ipmag.make_orthographic_map(200,20)

ipmag.plot_vgp(m, J_decs, J_incs, color='gray', edge='none', alpha=0.2)
```
